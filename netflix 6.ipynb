```python
df2['listed_in'].describe()
```

```python
plt.figure(figsize=(12, 6))
top_categories = df2['listed_in'].value_counts().head(15)
sns.barplot(y=top_categories.index, x=top_categories.values, palette='coolwarm')
plt.title('Top 15 Categories in listed_in')
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df2['rating'], kde=False, bins=5, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
country_counts = df2['country'].value_counts()
top_10_countries = country_counts.head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values)
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.title('Top 10 Countries with the Highest Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
df2['date_added'] = pd.to_datetime(df2['date_added'], errors='coerce')
df2['year_added'] = df2['date_added'].dt.year
year_counts = df2['year_added'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=year_counts.index, y=year_counts.values, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Number of Movies Added')
plt.title('Movies Added Per Year')
plt.xticks(rotation=45)
plt.show()
```
